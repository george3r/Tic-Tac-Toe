```python
# Step 6: Write a function that returns a boolean indicating whether a space on the board is freely available.

def space_check(board, position):
    
    return board[int(position)] not in ['X','O']
```

```python
# Step 7: Write a function that checks if the board is full and returns a boolean value. True if full, False otherwise.

def full_board_check(board):
    
    global board_full
    
    for i in board:
        if i not in ['X','O']: 
            board_full = False
            break
        else: board_full = True
            
    return board_full
```

```python
# Step 8: Write a function that asks for a player's next position (as a number 1-9) and then uses the function 
# from step 6 to check if it's a free position. If it is, then return the position for later use.

def player_choice(board):
    
    global position
    position = "0"
    
    while position not in str(list(range(1,10))) or position.isdigit() == False or space_check(board,int(position)) ==  False:       
        position = input("Please choose an unoccupied marker position between 1 & 9: ")

```

```python
# Step 9: Write a function that asks the player if they want to play again and returns a boolean True if they do want 
# to play again.

def replay():
    global play_again
    play_again = "wrong"
    
    while play_again not in ["Y","N"]:
        play_again = input('Would you like to play (Y/N)?  ').upper()
    return play_again == "Y"
```

```python
def choose_names():    
    global names_chosen
    global player1
    global player2
    if names_chosen == False:
        player1 = input('Please choose a name for player 1: ')
        player2 = input('please choose a name for player 2: ')
        names_chosen = True
```

```python
def stats():
   
    print('\n')
    print(f'{player1} has won {player1_won} games')
    print(f'{player2} has won {player2_won} games')
    print(f'{cats_games} games have been tied')
    print('\n')
```

```python
print('Welcome to Tic Tac Toe!')

names_chosen = False
player1_won = 0
player2_won = 0
cats_games = 0

# while game on: 
while replay() == True:

    initial_board = ['O','1','2','3','4','5','6','7','8','9']
    board = ['O',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    odd_turn = True
    
    display_board(initial_board)
    display_board(board)

    choose_names()
    choose_first()
    player_input()
    
    # Turns:
    while win_check(board,player1_mkr) == False and win_check(board,player2_mkr) == False and full_board_check(board) == False:
        
        #Odd Turn: 
        while odd_turn == True:
            player_choice(board)
            clear_output()
            place_marker(board,player1_mkr,int(position))
            display_board(board)
            odd_turn = False
            break
        # check for win con: 
        if win_check(board,player1_mkr) == True:
            print('\n')
            print(f'Congrats!!! {starting_player} ({player1_mkr}) won!!!')
            if starting_player == player1:
                player1_won += 1
            elif starting_player == player2:
                player2_won += 1
            stats()
            break
        elif full_board_check(board) == True:
            cats_games += 1
            print('\n')
            print('It is a CATS game')
            stats()
            break    
            
        # Even turn: 
        while odd_turn == False:
            player_choice(board)
            clear_output()
            place_marker(board,player2_mkr,int(position))
            display_board(board)
            odd_turn = True
            break
        # check for win con: 
        if win_check(board,player2_mkr) == True:
            print('\n')
            print(f'Congrats!!! {second_player} ({player2_mkr}) won!!!')
            if second_player == player1:
                player1_won += 1
            elif second_player == player2:
                player2_won += 1
            stats()
            break
        elif full_board_check(board) == True:
            cats_games += 1
            print('\n')
            print('It is a CATS game')
            stats()
            break
```
